#### Import necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Function to read and display the original image

```python
def read_and_display_image(file_path):
    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
    show_img_plt(img, 'Original Image')
```

#### Display image using matplotlib

```python
def show_img_plt(img, title='', c_map='gray', fig_height=8, fig_width=6):
    plt.figure(figsize=(fig_height, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map)
    plt.title(title)
    plt.show()
```

#### Display histogram using matplotlib

```python
def display_hist_plt(img, bins=256, img_range=(0, 256), title='Histogram'):
    plt.hist(img.flat, bins=bins, range=img_range)
    plt.title(title)
    plt.show()
```

#### Enhances the input image using Contrast Limited Adaptive Histogram Equalization (CLAHE)

```python
def enhance_image_with_CLAHE(img):
    # Create CLAHE object
    clahe = cv2.createCLAHE(clipLimit = 8.0, tileGridSize = (32, 32))
    # Apply CLAHE to the image
    enhanced_img = clahe.apply(img)
    return enhanced_img
```

### 1. Display Original Image and its Histogram

```python
file_path = 'sample.jpg'
```

```python
read_and_display_image(file_path)
display_hist_plt(cv2.imread(file_path, cv2.IMREAD_GRAYSCALE))
```

### 2. Read the original image

```python
original_img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
```

### 3. Enhance original image using CLAHE Algorithm

```python
enhanced_img = enhance_image_with_CLAHE(original_img)
```

### 4. Display the enhanced image

```python
show_img_plt(enhanced_img, title='Enhanced Image')
```

### 5. Display the histogram of the enhanced image

```python
display_hist_plt(enhanced_img)
```
